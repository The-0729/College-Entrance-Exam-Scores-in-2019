```python
# Add a columns 'A01' (a combination of subject)
df.loc[(df['Ma Ngoai Ngu'] == 'N1') & (df['Vat Ly'] != -1) & (df['Toan Hoc'] != -1), 'A01'] = df['Toan Hoc'] + df['Vat Ly'] + df['Ngoai Ngu']
df.loc[(df['Ma Ngoai Ngu'] != 'N1') | (df['Vat Ly'] == -1) | (df['Toan Hoc'] == -1), 'A01'] = 'Khong Thi' 
```

```python
# Take a look of the new column
khoi_a01 = df[df['A01'] != 'Khong Thi']
khoi_a01 = khoi_a01.sort_values('A01', ascending=False)
khoi_a01.head()
```

```python
# Add a columns 'A00' (a combination of subject)
df.loc[(df['Hoa Hoc'] != -1) & (df['Vat Ly'] != -1) & (df['Toan Hoc'] != -1), 'A00'] = df['Toan Hoc'] + df['Vat Ly'] + df['Hoa Hoc']
df.loc[(df['Hoa Hoc'] == -1) | (df['Vat Ly'] == -1) | (df['Toan Hoc'] == -1), 'A00'] = 'Khong Thi' 
```

```python
# Add a columns 'B00' (a combination of subject)
df.loc[(df['Hoa Hoc'] != -1) & (df['Sinh Hoc'] != -1) & (df['Toan Hoc'] != -1), 'B00'] = df['Toan Hoc'] + df['Sinh Hoc'] + df['Hoa Hoc']
df.loc[(df['Hoa Hoc'] == -1) | (df['Sinh Hoc'] == -1) | (df['Toan Hoc'] == -1), 'B00'] = 'Khong Thi' 
```

```python
# Add a columns 'D01' (a combination of subject)
df.loc[(df['Ma Ngoai Ngu'] == 'N1') & (df['Ngu Van'] != -1) & (df['Toan Hoc'] != -1), 'D01'] = df['Toan Hoc'] + df['Ngu Van'] + df['Ngoai Ngu']
df.loc[(df['Ma Ngoai Ngu'] != 'N1') | (df['Ngu Van'] == -1) | (df['Toan Hoc'] == -1), 'D01'] = 'Khong Thi' 
```

```python
# Add a columns C00' (a combination of subject)
df.loc[(df['Ngu Van'] != -1) & (df['Lich Su'] != -1) & (df['Dia Ly'] != -1), 'C00'] = df['Lich Su'] + df['Ngu Van'] + df['Dia Ly']
df.loc[(df['Ngu Van'] == -1) | (df['Lich Su'] == -1) | (df['Dia Ly'] == -1), 'C00'] = 'Khong Thi'
```

```python
# Add a columns 'D07' (a combination of subject)
df.loc[(df['Ma Ngoai Ngu'] == 'N1') & (df['Hoa Hoc'] != -1) & (df['Toan Hoc'] != -1), 'D07'] = df['Toan Hoc'] + df['Hoa Hoc'] + df['Ngoai Ngu']
df.loc[(df['Ma Ngoai Ngu'] != 'N1') | (df['Hoa Hoc'] == -1) | (df['Toan Hoc'] == -1), 'D07'] = 'Khong Thi' 
```

```python
# Add a columns 'C03' (a combination of subject)
df.loc[(df['Ngu Van'] != -1) & (df['Lich Su'] != -1) & (df['Toan Hoc'] != -1), 'C03'] = df['Toan Hoc'] + df['Lich Su'] + df['Ngu Van']
df.loc[(df['Ngu Van'] == -1) | (df['Lich Su'] == -1) | (df['Toan Hoc'] == -1), 'C03'] = 'Khong Thi' 
```

```python
# Add a columns 'Khac' (different from the above)
df.loc[(df['Ma Ngoai Ngu'] != 'N1') & (df['Ma Ngoai Ngu'] != 'Khong Thi') & (df['Ngu Van'] != -1) & (df['Toan Hoc'] != -1), 'Khac'] = df['Toan Hoc'] + df['Ngu Van'] + df['Ngoai Ngu']
df['Khac'] = df['Khac'].fillna('Khong Thi')
```

```python
# Get rid of rows with value 'Khong Thi' in the whole row
df_max = df[(df['A00'] != 'Khong Thi') | (df['A01'] != 'Khong Thi') | (df['B00'] != 'Khong Thi') | (df['D01'] != 'Khong Thi') | (df['Khac'] != 'Khong Thi') | (df['C00'] != 'Khong Thi') | (df['D07'] != 'Khong Thi') | (df['C03'] != 'Khong Thi')]
```

```python
# Select specific columns of the dataframe above
df_max1 = df_max[['A01', 'A00', 'B00', 'D01', 'C00', 'D07', 'C03', 'Khac']]
df_max1
```

```python
# Add the Max Score column for each row
df_max1 = df_max1.replace('Khong Thi', 0)
df_max1 = df_max1.assign(max=df_max1.values.max(1))
```

```python
df_max1
```

```python
# Add the 'khoi' (a combination of subjects) column (Most optimal choice) for each row
df_max1.loc[df_max1['max'] == df_max1['A01'], 'khoi'] = 'A01'
df_max1.loc[df_max1['max'] == df_max1['A00'], 'khoi'] = 'A00'
df_max1.loc[df_max1['max'] == df_max1['B00'], 'khoi'] = 'B00'
df_max1.loc[df_max1['max'] == df_max1['D01'], 'khoi'] = 'D01'
df_max1.loc[df_max1['max'] == df_max1['D07'], 'khoi'] = 'D07'
df_max1.loc[df_max1['max'] == df_max1['C03'], 'khoi'] = 'C03'
df_max1.loc[df_max1['max'] == df_max1['C00'], 'khoi'] = 'C00'
df_max1.loc[df_max1['max'] == df_max1['Khac'], 'khoi'] = 'Khac'
```

```python
# Count index in column 'khoi'
df_max1['khoi'].value_counts()
```

```python
# Round the score to .0 or .5
df_max1['max'] = df_max1['max'].apply(lambda x: int(x*2+0.99)/2)
```

```python
# Concatenate 2 dataframe into a result dataframe and return a csv file
df_max1 = df_max1[df_max1.columns.difference(df_max.columns)]
df_total = pd.concat([df_max, df_max1], axis = 1)
```
